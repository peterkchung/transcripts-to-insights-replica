```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify
```

```python
url = 'https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average'


tables = pd.read_html(url)
df = tables[1]

df['member'] = df['Symbol'] + " - " + df['Company']

df['weight'] = df['Index weighting'].str.strip("%").astype(float).div(100)

df.to_csv('output/DJIA.csv')

df.head()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
squarify.plot(sizes=df['weight'], label=df['member'], alpha=.8)
plt.axis('off')
plt.show()
```
